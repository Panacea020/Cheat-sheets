# Dealing with missing values

```python
missing_data = dataset.isnull()
missing_data.head()
```

```python
cols_with_missing = [col for col in dataset.columns if dataset[col].isnull().any()]
print("Number of columns with missing values : " , len(cols_with_missing))
```

# replace missing value with a mean / drop the rows with null

```python
#replace with mean
```

```python
dataset['China(CNY)'].fillna(int(dataset['China(CNY)'].mean()), inplace=True)
dataset['Turkey(TRY)'].fillna(int(dataset['Turkey(TRY)'].mean()), inplace=True)
dataset['Vietnam(VND)'].fillna(int(dataset['Vietnam(VND)'].mean()), inplace=True)
dataset['Egypt(EGP)'].fillna(int(dataset['Egypt(EGP)'].mean()), inplace=True)
dataset.isnull().sum()
```

# PLOTS DATASET

```python
import matplotlib.pyplot as plt 
```

```python
plt.plot(dataset["South Africa(ZAR)"] )
plt.xlabel("Months")
plt.ylabel("Price(ZAR)")
plt.title("Monthly Gold Price in South Africa ");
```
